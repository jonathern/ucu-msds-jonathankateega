### Question 1
#### UBOS Multi-District Population & Growth Forecast

Given the following dataset (Population estimates in thousands for 10 years):
Kampala = [1200, 1250, 1300, 1350, 1420, 1500, 1580, 1650, 1720, 1800]
Wakiso = [950, 1000, 1070, 1150, 1220, 1300, 1390, 1480, 1570, 1670]
Gulu = [320, 330, 345, 360, 375, 390, 410, 430, 455, 480]

Attempt the following tasks:
1. Store the data for each district in a NumPy array.
2. Use the statistics module to compute mean, median, variance, and standard
deviation for each district.
3. Generate a Fibonacci sequence of length 5 to project the next 5 years of growth.
4. Compare the variance of actual vs projected data.
5. Plot actual vs projected populations using Matplotlib.

```python
kampala = [1200, 1250, 1300, 1350, 1420, 1500, 1580, 1650, 1720, 1800]
wakiso = [950, 1000, 1070, 1150, 1220, 1300, 1390, 1480, 1570, 1670]
gulu = [320, 330, 345, 360, 375, 390, 410, 430, 455, 480]
```

```python
"""
1.1. Store the data for each district in a NumPy array.
"""

import numpy as np

kampala_arr = np.array(kampala)
wakiso_arr = np.array(wakiso)
gulu_arr = np.array(gulu)

print(f"The Kampala numpy array is {kampala_arr}.")
print(f"The Wakiso numpy array is {wakiso_arr}.")
print(f"The Gulu numpy array is {gulu_arr}.")
```

```python
"""
1.2. Use the statistics module to compute mean, median, variance, and standard
deviation for each district.
"""

def calculate_stats(arr_name, arr):
    mean = np.mean(arr)
    median = np.median(arr)
    variance = np.var(arr)
    standard_deviation = np.std(arr)
    print(f"{arr_name}\nMean: {mean}\nMedian: {median}\nVariance: {variance}\nStandard Deviation: {standard_deviation}\n")
    
calculate_stats("Kampala", kampala_arr)
calculate_stats("Wakiso", wakiso_arr)
calculate_stats("Gulu", gulu_arr)
```

```python
"""
1.3. Generate a Fibonacci sequence of length 5 to project the next 5 years of growth.
"""

def fibonacci_projection(arr, num_to_add):

    # Create a copy of the original array
    projection = arr.tolist()
    
    for _ in range(num_to_add):
        # Get the last two numbers using negative indexing
        last_num = projection[-1] 
        second_last_num = projection[-2]

        # Calculate the next Fibonacci-like number
        next_fib = last_num + second_last_num

        # Append the new number to the end of the projection array
        projection.append(next_fib)

    projection_arr = np.array(projection)
    return projection_arr

# print(f"Original array: {kampala_arr}")

# # Continue the sequence by adding 5 new values
# updated_array = fibonacci_projection(kampala_arr, 5)
# print(f"Updated array: {updated_array}")

city_pops = {
    "Kampala": kampala_arr,
    "Wakiso": wakiso_arr,
    "Gulu": gulu_arr
}

for city in city_pops.keys():

    city_array = city_pops[city]
    
    print(f"Original array for {city}: {city_array}")

    # Continue the sequence by adding 5 new values
    updated_array = fibonacci_projection(city_array, 5)
    print(f"Updated array for {city}: {updated_array}\n")
```

```python
"""
1.4. Compare the variance of actual vs projected data.
"""

for city in city_pops.keys():

    city_array = city_pops[city]
    
    # variance in the original array
    original_variance = np.var(city_array)
    print(f"The variance in the original array for {city}: {original_variance}")

    # variance in the last 5 values in the updated array
    updated_array = fibonacci_projection(city_array, 5)
    variance_updated = np.var(updated_array[-5:])
    print(f"The variance in the projected data for {city} is: {variance_updated}")
    
    # difference in the variance
    variance_diff = original_variance - variance_updated
    print(f"The difference in variance for {city} is: {variance_diff}\n")
    
```

```python
"""
1.5. Plot actual vs projected populations using Matplotlib.
"""

import matplotlib.pyplot as plt

for city in city_pops.keys():

    city_array = city_pops[city]
    
    # plot the original array
    plt.plot(city_array, label=city)
    
    # plot the last 5 values in the updated array
    updated_array = fibonacci_projection(city_array, 5)
    plt.plot(updated_array[-5:], label=f"{city} (projected)")

    plt.xlabel("Year")
    plt.ylabel("Population")
    plt.legend()
    plt.show()
    plt.close()
```

### Question 2
#### Solar Micro-Grid Simulation

Equations: 3x + 2y = D1 and 4x + y = D2, where x = solar panels, y = batteries.

Attempt the following tasks:
1. Prompt the user to enter daily demand values D1 and D2.
2. Use scipy.linalg.solve() to calculate energy from solar panels and batteries.
3. Repeat for 7 days, storing results in a NumPy array.
4. Use statistics to compute mean, variance, and standard deviation of solar vs battery usage.
5. Plot daily solar vs battery usage.

```python
"""
2.1. Prompt the user to enter daily demand values D1 and D2.
2.2. Use scipy.linalg.solve() to calculate energy from solar panels and batteries.
2.3. Repeat for 7 days, storing results in a NumPy array.
"""

from scipy.linalg import solve
import statistics as stats

# Coefficient matrix for the equations:
# 3x + 2y = D1
# 4x +  y = D2
A = np.array([[3, 2],
              [4, 1]])

# To store results for 7 days
results = []

print("Enter daily demand values for 7 days:")

for day in range(1, 8):
    D1 = float(input(f"Day {day} - Enter D1: "))
    D2 = float(input(f"Day {day} - Enter D2: "))

    # Solve for [x, y]
    b = np.array([D1, D2])
    x, y = solve(A, b)

    # Store results
    results.append([x, y])

# Convert to NumPy array
results = np.array(results)
solar = results[:, 0]
battery = results[:, 1]
```

```python
"""
2.4. Use statistics to compute mean, variance, and standard deviation of solar vs battery usage.
"""

mean_solar = np.mean(solar)
mean_battery = np.mean(battery)
var_solar = np.var(solar)
var_battery = np.var(battery)
std_solar = np.std(solar)
std_battery = np.std(battery)

print("\n--- Energy Statistics ---")
print(f"Mean Solar Usage: {mean_solar:.2f}")
print(f"Mean Battery Usage: {mean_battery:.2f}")
print(f"Variance (Solar): {var_solar:.2f}")
print(f"Variance (Battery): {var_battery:.2f}")
print(f"Std Dev (Solar): {std_solar:.2f}")
print(f"Std Dev (Battery): {std_battery:.2f}")
```

```python
"""
2.5. Plot daily solar vs battery usage.
"""

days = np.arange(1, 8)
plt.figure(figsize=(8, 5))
plt.plot(days, solar, marker='o', label='Solar Panels (x)')
plt.plot(days, battery, marker='s', label='Batteries (y)')
plt.title('Daily Solar vs Battery Energy Usage')
plt.xlabel('Day')
plt.ylabel('Energy Units')
plt.grid(True)
plt.legend()
plt.show()
```

### Question 4
#### Weather & Agriculture Analysis

Rainfall data (mm) for Kampala across 12 months: [120, 140, 180, 200, 220, 180, 90, 70, 60, 100, 110, 130]

Attempt the following tasks:

1. Store rainfall data in a NumPy array.
2. Compute mean rainfall and classify months as 'Good for maize' (>150mm) or 'Drought risk' (<150mm).
3. Simulate Gulu rainfall with different values.
4. Compute cosine similarity of Kampala vs Gulu rainfall trends using math.cos().
5. Plot both regions' rainfall in one chart.

```python
"""
4.1. Store rainfall data in a NumPy array.
"""

kampala_rainfall = np.array([120, 140, 180, 200, 220, 180, 90, 70, 60, 100, 110, 130])
```

```python
"""
4.2. Compute mean rainfall and classify months as 'Good for maize' (>150mm) or 'Drought risk' (<150mm).
"""

mean_kampala = np.mean(kampala_rainfall)
print(f"Mean monthly rainfall (Kampala): {mean_kampala:.2f} mm\n")
```

```python
classification = [
    "Good for maize" if val > 150 else "Drought risk"
    for val in kampala_rainfall
]

for i, (val, cls) in enumerate(zip(kampala_rainfall, classification), start=1):
    print(f"Month {i}: {val} mm  {cls}")
```

```python
"""
4.3. Simulate Gulu rainfall with different values.
"""

np.random.seed(42)  # for reproducibility
gulu_rainfall = kampala_rainfall + np.random.randint(-30, 30, size=12)

print("\nSimulated Gulu rainfall (mm):", gulu_rainfall)
```

```python
"""
4.4. Compute cosine similarity of Kampala vs Gulu rainfall trends using math.cos().
"""

import math

# Cosine similarity = (AB) / (||A|| * ||B||)
dot_product = np.dot(kampala_rainfall, gulu_rainfall)
magnitude_kampala = np.linalg.norm(kampala_rainfall)
magnitude_gulu = np.linalg.norm(gulu_rainfall)

cosine_similarity = dot_product / (magnitude_kampala * magnitude_gulu)
angle_radians = math.acos(cosine_similarity)  # using math.cos() inverse
angle_degrees = math.degrees(angle_radians)


print(f"\nCosine Similarity (Kampala vs Gulu): {cosine_similarity:.4f}")
print(f"Angle between trends: {angle_degrees:.2f}")
```

```python
"""
4.5. Plot both regions' rainfall in one chart.
"""

months = np.arange(1, 13)

plt.figure(figsize=(9, 5))
plt.plot(months, kampala_rainfall, marker='o', label='Kampala', color='teal')
plt.plot(months, gulu_rainfall, marker='s', label='Gulu', color='orange')
plt.title("Monthly Rainfall Comparison: Kampala vs Gulu")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
"""
5.2. Compute daily revenues at UGX 2000 per passenger.
"""

fare = 2000  # UGX per passenger
revenues = passengers * fare
print("Daily Revenues (UGX):", revenues)
```

```python
"""
5.3. Use statistics to analyze mean, variance, and std deviation.
"""

mean_rev = stats.mean(revenues)
variance_rev = stats.variance(revenues)
std_rev = stats.stdev(revenues)

print(f"\nMean Revenue: {mean_rev:,.0f} UGX")
print(f"Variance: {variance_rev:,.0f}")
print(f"Standard Deviation: {std_rev:,.0f}")
```

```python
"""
5.4. Use scipy.linalg.solve to model a simple supply-demand system.
"""

# 2x + y = 100  (supply constraint)
# x - y = 10    (demand imbalance)
# -------------------------------

A = np.array([[2, 1],
              [1, -1]])
b = np.array([100, 10])

solution = solve(A, b)
supply, demand = solution
print(f"\nSupply = {supply:.2f}, Demand = {demand:.2f}")
```
