Question 1: UBOS Multi-District Population & Growth Forecast
Given the following dataset (Population estimates in thousands for 10 years):
Kampala = [1200, 1250, 1300, 1350, 1420, 1500, 1580, 1650, 1720, 1800]
Wakiso = [950, 1000, 1070, 1150, 1220, 1300, 1390, 1480, 1570, 1670]
Gulu = [320, 330, 345, 360, 375, 390, 410, 430, 455, 480]

Attempt the following tasks:
1.1. Store the data for each district in a NumPy array.
1.2. Use the statistics module to compute mean, median, variance, and standard
deviation for each district.
1.3. Generate a Fibonacci sequence of length 5 to project the next 5 years of growth.
1.4. Compare the variance of actual vs projected data.
1.5. Plot actual vs projected populations using Matplotlib.

```python
kampala = [1200, 1250, 1300, 1350, 1420, 1500, 1580, 1650, 1720, 1800]
wakiso = [950, 1000, 1070, 1150, 1220, 1300, 1390, 1480, 1570, 1670]
gulu = [320, 330, 345, 360, 375, 390, 410, 430, 455, 480]
```

```python
"""
1.1. Store the data for each district in a NumPy array.
"""

import numpy as np

kampala_arr = np.array(kampala)
wakiso_arr = np.array(wakiso)
gulu_arr = np.array(gulu)

print(f"The Kampala numpy array is {kampala_arr}.")
print(f"The Wakiso numpy array is {wakiso_arr}.")
print(f"The Gulu numpy array is {gulu_arr}.")
```

```python
"""
1.2. Use the statistics module to compute mean, median, variance, and standard
deviation for each district.
"""

def calculate_stats(arr_name, arr):
    mean = np.mean(arr)
    median = np.median(arr)
    variance = np.var(arr)
    standard_deviation = np.std(arr)
    print(f"{arr_name}\nMean: {mean}\nMedian: {median}\nVariance: {variance}\nStandard Deviation: {standard_deviation}\n")
    
calculate_stats("Kampala", kampala_arr)
calculate_stats("Wakiso", wakiso_arr)
calculate_stats("Gulu", gulu_arr)
```

```python
"""
1.3. Generate a Fibonacci sequence of length 5 to project the next 5 years of growth.
"""

def fibonacci_projection(arr, num_to_add):

    # Create a copy of the original array
    projection = arr.tolist()
    
    for _ in range(num_to_add):
        # Get the last two numbers using negative indexing
        last_num = projection[-1] 
        second_last_num = projection[-2]

        # Calculate the next Fibonacci-like number
        next_fib = last_num + second_last_num

        # Append the new number to the end of the projection array
        projection.append(next_fib)

    projection_arr = np.array(projection)
    return projection_arr

# print(f"Original array: {kampala_arr}")

# # Continue the sequence by adding 5 new values
# updated_array = fibonacci_projection(kampala_arr, 5)
# print(f"Updated array: {updated_array}")

city_pops = {
    "Kampala": kampala_arr,
    "Wakiso": wakiso_arr,
    "Gulu": gulu_arr
}

for city in city_pops.keys():

    city_array = city_pops[city]
    
    print(f"Original array for {city}: {city_array}")

    # Continue the sequence by adding 5 new values
    updated_array = fibonacci_projection(city_array, 5)
    print(f"Updated array for {city}: {updated_array}\n")
```

```python
"""
1.4. Compare the variance of actual vs projected data.
"""

for city in city_pops.keys():

    city_array = city_pops[city]
    
    # variance in the original array
    original_variance = np.var(city_array)
    print(f"The variance in the original array for {city}: {original_variance}")

    # variance in the last 5 values in the updated array
    updated_array = fibonacci_projection(city_array, 5)
    variance_updated = np.var(updated_array[-5:])
    print(f"The variance in the projected data for {city} is: {variance_updated}")
    
    # difference in the variance
    variance_diff = original_variance - variance_updated
    print(f"The difference in variance for {city} is: {variance_diff}\n")
    
```

```python
"""
1.5. Plot actual vs projected populations using Matplotlib.
"""

import matplotlib.pyplot as plt

for city in city_pops.keys():

    city_array = city_pops[city]
    
    # plot the original array
    plt.plot(city_array, label=city)
    
    # plot the last 5 values in the updated array
    updated_array = fibonacci_projection(city_array, 5)
    plt.plot(updated_array[-5:], label=f"{city} (projected)")

    plt.xlabel("Year")
    plt.ylabel("Population")
    plt.legend()
    plt.show()
    plt.close()
```

Question 2: Solar Micro-Grid Simulation
Equations: 3x + 2y = D1 and 4x + y = D2, where x = solar panels, y = batteries.
Attempt the following tasks:
2.1. Prompt the user to enter daily demand values D1 and D2.
2.2. Use scipy.linalg.solve() to calculate energy from solar panels and batteries.
2.3. Repeat for 7 days, storing results in a NumPy array.
2.4. Use statistics to compute mean, variance, and standard deviation of solar vs
battery usage.
2.5. Plot daily solar vs battery usage.

```python
"""
2.1. Prompt the user to enter daily demand values D1 and D2.
2.2. Use scipy.linalg.solve() to calculate energy from solar panels and batteries.
2.3. Repeat for 7 days, storing results in a NumPy array.
"""

from scipy.linalg import solve
import statistics as stats

# Coefficient matrix for the equations:
# 3x + 2y = D1
# 4x +  y = D2
A = np.array([[3, 2],
              [4, 1]])

# To store results for 7 days
results = []

print("Enter daily demand values for 7 days:")

for day in range(1, 8):
    D1 = float(input(f"Day {day} - Enter D1: "))
    D2 = float(input(f"Day {day} - Enter D2: "))

    # Solve for [x, y]
    b = np.array([D1, D2])
    x, y = solve(A, b)

    # Store results
    results.append([x, y])

# Convert to NumPy array
results = np.array(results)
solar = results[:, 0]
battery = results[:, 1]
```

```python
"""
2.4. Use statistics to compute mean, variance, and standard deviation of solar vs battery usage.
"""

mean_solar = np.mean(solar)
mean_battery = np.mean(battery)
var_solar = np.var(solar)
var_battery = np.var(battery)
std_solar = np.std(solar)
std_battery = np.std(battery)

print("\n--- Energy Statistics ---")
print(f"Mean Solar Usage: {mean_solar:.2f}")
print(f"Mean Battery Usage: {mean_battery:.2f}")
print(f"Variance (Solar): {var_solar:.2f}")
print(f"Variance (Battery): {var_battery:.2f}")
print(f"Std Dev (Solar): {std_solar:.2f}")
print(f"Std Dev (Battery): {std_battery:.2f}")
```

```python
"""
2.5. Plot daily solar vs battery usage.
"""

days = np.arange(1, 8)
plt.figure(figsize=(8, 5))
plt.plot(days, solar, marker='o', label='Solar Panels (x)')
plt.plot(days, battery, marker='s', label='Batteries (y)')
plt.title('Daily Solar vs Battery Energy Usage')
plt.xlabel('Day')
plt.ylabel('Energy Units')
plt.grid(True)
plt.legend()
plt.show()
```
